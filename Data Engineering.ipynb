```python
import pandas as pd
```

GEN datasets

```python
gen_ccgt = pd.read_csv("Individual Assignment/Gen/gen_ccgt.csv")
gen_coal = pd.read_csv("Individual Assignment/Gen/gen_coal.csv")
gen_hydro = pd.read_csv("Individual Assignment/Gen/gen_hydro.csv")
gen_nuclear = pd.read_csv("Individual Assignment/Gen/gen_nuclear.csv")
gen_solar_pv = pd.read_csv("Individual Assignment/Gen/gen_solar_pv.csv")
gen_solar_th = pd.read_csv("Individual Assignment/Gen/gen_solar_th.csv")
gen_total = pd.read_csv("Individual Assignment/Gen/gen_total.csv")
gen_wind = pd.read_csv("Individual Assignment/Gen/gen_wind.csv")
```

```python
gen_ccgt['gen_type'] = 'CCGT'
gen_coal['gen_type'] = 'Coal'
gen_hydro['gen_type'] = 'Hydro'
gen_nuclear['gen_type'] = 'Nuclear'
gen_solar_pv['gen_type'] = 'Solar PV'
gen_solar_th['gen_type'] = 'Solar Th'
gen_total['gen_type'] = 'Total'
gen_wind['gen_type'] = 'Wind'
```

```python
# Concatenate all DataFrames into one
all_gen = pd.concat(
    [gen_ccgt, gen_coal, gen_hydro, gen_nuclear, gen_solar_pv, gen_solar_th, gen_total, gen_wind]
)

# Reset the index of the merged DataFrame
all_gen.reset_index(drop=True, inplace=True)
```

```python
all_gen
```

```python
all_gen['gen_type'].unique()
```

```python
# Reset the index of the merged DataFrame
all_gen.reset_index(drop=True, inplace=True)

# Save the merged dataset to a CSV file
all_gen.to_csv('gen.csv', index=False)
```
